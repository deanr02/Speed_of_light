```python
import numpy as np
import matplotlib.pyplot as plt

# Historical measurements of the speed of light (in km/s) with years
historical_data = [  
    (1849, 315300, 500),   # Fizeau
    (1862, 298000, 500),    #Foucalt
    (1874, 300500, 200),    #Cornu
    (1879, 299910, 50),    # Michelson
    (1926, 299798, 4),     # Michelson
    (1950, 299792.5, 1), # Essen
    (1973, 299792.457, 0.001) # Laser interferometry
]

# Modern accepted value (fixed by definition in m/s converted to km/s)
modern_value = 299792.458

# Extract years, measurements, and uncertainties
years, measurements, uncertainties = zip(*historical_data)

# Create the box plot
plt.figure(figsize=(8, 6))
#plt.scatter(measurements, vert=True, widths=10, patch_artist=True, boxprops=dict(facecolor='lightblue'))

# Add individual data points with error bars
plt.errorbar(years, measurements, yerr=uncertainties, fmt='o', color='black', label='Historical Measurements')

# Add a horizontal line for the modern value
plt.axhline(modern_value, color='red', linestyle='--', label='Modern Value')

# Labels and title
plt.ylabel("Speed of Light (km/s)")
plt.xlabel("Year")
plt.title("Historical Measurements of the Speed of Light")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```
